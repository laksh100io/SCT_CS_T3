```python
# -*- coding: utf-8 -*-
"""Welcome To Colab

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/notebooks/intro.ipynb
"""

import streamlit as st
import re

# Function to check password strength
def check_password_strength(password):
    strength = 0
    criteria = {
        "Length (>= 8 chars)": len(password) >= 8,
        "Contains uppercase": re.search(r"[A-Z]", password) is not None,
        "Contains lowercase": re.search(r"[a-z]", password) is not None,
        "Contains digit": re.search(r"[0-9]", password) is not None,
        "Contains special char": re.search(r"[!@#$%^&*(),.?\":{}|<>]", password) is not None
    }

    for passed in criteria.values():
        if passed:
            strength += 1

    if strength <= 2:
        label = "‚ùå Weak"
        color = "red"
    elif strength == 3 or strength == 4:
        label = "‚ö†Ô∏è Moderate"
        color = "orange"
    else:
        label = "‚úÖ Strong"
        color = "green"

    return label, color, criteria

# --- Streamlit UI ---
st.set_page_config(page_title="Password Strength Checker", page_icon="üîê")

st.title("üîê Password Strength Checker")
st.write("Enter a password to check its strength based on common criteria.")

# Password input field
password = st.text_input("Enter your password", type="password")

if password:
    label, color, details = check_password_strength(password)

    # Strength result
    st.markdown(f"### Strength: <span style='color:{color}'>{label}</span>", unsafe_allow_html=True)

    # Detailed breakdown
    st.markdown("#### Criteria:")
    for rule, passed in details.items():
        icon = "‚úÖ" if passed else "‚ùå"
        st.markdown(f"- {icon} {rule}")
```
